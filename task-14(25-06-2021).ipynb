```python
import pandas as pd
import numpy as np
```

```python
titanic = pd.read_csv('https://raw.githubusercontent.com/nagamounika5/Datasets/master/titanic.csv')
titanic
```

```python
titanic.isnull().sum()
```

```python
titanic.isnull().sum().sum()
```

```python
# Relace null values in 'Age' column with ffill()
titanic['Age'].fillna(method = "ffill", inplace = True)
titanic
```

```python
# Replace null values in cabin and embarked colmn wiht most occured value
titanic['Cabin'].fillna(titanic['Cabin'].mode().iloc[0], inplace = True)
titanic['Embarked'].fillna(titanic['Embarked'].mode().iloc[0], inplace = True)
titanic
```

```python
titanic.isnull().sum()
```

```python
#Task 2:
#After handiling missing values. convert cabin , embarked columns into numberical columns

from sklearn.preprocessing import LabelEncoder
nu=LabelEncoder()
```

```python
titanic['Cabin'] = nu.fit_transform(titanic['Cabin'])
titanic['Embarked'] = nu.fit_transform(titanic['Embarked'])
titanic
```
